```python
robot_art = r"""
      0: {head_name}
      Is available: {head_status}
      Attack: {head_attack}                              
      Defense: {head_defense}
      Energy consumption: {head_energy_consumption}
              ^
              |                  |1: {weapon_name}
              |                  |Is available: {weapon_status}
     ____     |    ____          |Attack: {weapon_attack}
    |oooo|  ____  |oooo| ------> |Defense: {weapon_defense}
    |oooo| '    ' |oooo|         |Energy consumption: {weapon_energy_consumption}
    |oooo|/\_||_/\|oooo|          
    `----' / __ \  `----'           |2: {left_arm_name}
   '/  |#|/\/__\/\|#|  \'           |Is available: {left_arm_status}
   /  \|#|| |/\| ||#|/  \           |Attack: {left_arm_attack}
  / \_/|_|| |/\| ||_|\_/ \          |Defense: {left_arm_defense}
 |_\/    O\=----=/O    \/_|         |Energy consumption: {left_arm_energy_consumption}
 <_>      |=\__/=|      <_> ------> |
 <_>      |------|      <_>         |3: {right_arm_name}
 | |   ___|======|___   | |         |Is available: {right_arm_status}
// \\ / |O|======|O| \  //\\        |Attack: {right_arm_attack}
|  |  | |O+------+O| |  |  |        |Defense: {right_arm_defense}
|\/|  \_+/        \+_/  |\/|        |Energy consumption: {right_arm_energy_consumption}
\__/  _|||        |||_  \__/        
      | ||        || |          |4: {left_leg_name} 
     [==|]        [|==]         |Is available: {left_leg_status}
     [===]        [===]         |Attack: {left_leg_attack}
      >_<          >_<          |Defense: {left_leg_defense}
     || ||        || ||         |Energy consumption: {left_leg_energy_consumption}
     || ||        || || ------> |
     || ||        || ||         |5: {right_leg_name}
   __|\_/|__    __|\_/|__       |Is available: {right_leg_status}
  /___n_n___\  /___n_n___\      |Attack: {right_leg_attack}
                                |Defense: {right_leg_defense}
                                |Energy consumption: {right_leg_energy_consumption}
                                
"""
```

```python
class Part:

  def __init__(self, name, attack_level, defense_level, energy_consumptions):
    self.name=name
    self.attack_level=attack_level
    self.defense_level=defense_level
    self.energy_consumptions=energy_consumptions

  def get_status_dict(self):
    formatted_name=self.name.replace(" ","_").lower()
    return {
        "{}_name".format(formatted_name): self.name.upper(),
        "{}_status".format(formatted_name): self.is_available(),
        "{}_attack".format(formatted_name): self.attack_level,
        "{}_defense".format(formatted_name): self.defense_level,
        "{}_energy_consumption".format(formatted_name): self.energy_consumptions,
    }

  def is_available(self):
    return not self.defense_level <= 0
```

```python
colors = {
    "black":'\x1b[90m',
    "blue":'\x1b[94m',
    "cyan":'\x1b[96m',
    "green":'\x1b[92m',
    "magenta":'\x1b[95m',
    "red":'\x1b[91m',
    "white":'\x1b[97m',
    "yellow":'\x1b[93m',
}

coins = {
         "head": 2,
         "weapon": 2,
         "left_arm": 1,
         "right_arm": 1,
         "left_leg": 1,
         "right_leg": 1,
}

class Robot:
  def __init__(self, name, color_code):
    self.name=name
    self.color_code=color_code
    self.energy = 100
    self.coin = 0
    self.tienda = 0
    self.parts = [
        Part("head", attack_level=10, defense_level=20, energy_consumptions=5),
        Part("weapon", attack_level=15, defense_level=20, energy_consumptions=10),
        Part("left_arm", attack_level=13, defense_level=30, energy_consumptions=10),
        Part("right_arm", attack_level=12, defense_level=30, energy_consumptions=10),
        Part("left_leg", attack_level=13, defense_level=30, energy_consumptions=15),
        Part("right_leg", attack_level=12, defense_level=30, energy_consumptions=15),
    ]
    
  def greet(self):    
    print("Hello, my name is", self.name)
  
  def print_energy(self):
    print("We have", self.energy, "percent energy left")

  def print_coin(self):
    print("we have", self.coin, "coins")
  
  def attack(self, enemy_robot, part_to_use, part_to_attack):
    enemy_robot.parts[part_to_attack].defense_level -= self.parts[part_to_use].attack_level
    self.energy -= self.parts[part_to_use].energy_consumptions

  def is_no(self):
    return self.energy < 0 
  
  def is_there_available_parts(self):
    for part in self.parts:
      if part.is_available() == True:
        return False
      return True

  def print_status(self):
    print(self.color_code)
    str_robot = robot_art.format(**self.get_part_status())
    self.greet()
    self.print_energy()
    self.print_coin()
    print(str_robot)
    print(colors["black"])

  def get_part_status(self):
    part_status = {}
    for part in self.parts:
      status_dict = part.get_status_dict()
      part_status.update(status_dict)
    return part_status
 
  def add_coins(self,part_to_attack):
    for coin_for in coins.keys():
      if coin_for == self.parts[part_to_attack].name:
        self.coin += coins[coin_for]

  def add_increase(self, decision, typo1):
    if typo1 == 1:
      amount = self.coin * 0.1
      percentage = int(self.parts[decision].defense_level * amount)
      self.parts[decision].defense_level += percentage
      self.coin = 0

    elif typo1 == 2:
      amount = self.coin * 0.1
      percentage = int(self.parts[decision].attack_level * amount)
      self.parts[decision].attack_level += percentage
      self.coin = 0
      
  def total_coins(self):
    return self.coin
  
  def recorrer_tienda(self):
    return self.tienda

def menu_parts():
  print("0). Head\n"
        "1). Weapon\n"
        "2). Left_arm\n" 
        "3). Right_arm\n" 
        "4). Left_leg\n" 
        "5). Right_leg\n")

def one_decision(current_robot):
  print("1). attack\n"
        "2). shop\n")
  process = int(input("enter the data you want to make: "))
  if process == 1:
    rot = 1
  elif process == 2:
    current_robot.tienda += 1
    print("1). defending\n"
          "2). attack\n")
    typo = int(input("choose the improvement you want to make: "))
    if typo == 1:
      menu_parts()
      typo1 = 1
      decision = int(input("write the part you want to upgrade: "))
      current_robot.add_increase(decision, typo1)
      current_robot.print_status()
    elif typo == 2:
      menu_parts()
      typo1 = 2
      decision = int(input("write the part you want to upgrade: "))
      current_robot.add_increase(decision, typo1)
      current_robot.print_status()

def build_robot():
  robot_name=input("Robot name:  ")
  choose_color()
  color = input("write the color: ").lower()
  color_code = colors[color]
  robot = Robot(robot_name, color_code)
  robot.print_status()
  return robot

def choose_color():
  print(colors["black"] + "1). Black\n",
        colors["blue"] + "2). Blue\n",
        colors["cyan"] + "3). Cyan\n",
        colors["green"] + "4). Green\n",
        colors["magenta"] + "5). Magenta\n",
        colors["red"] + "6). Red\n",
        colors["black"] + "7). White\n",
        colors["yellow"] + "8).Â Yellow\n",)
  
def store_on(current_robot):
   if current_robot.total_coins() > 0 and current_robot.recorrer_tienda() < 2:
      one_decision(current_robot)

def play():
  print("Welcome to the game")
  print("Data for player 1")
  robot_one = build_robot()
  print("Data for player 2")
  robot_two = build_robot()
  print("---------------------------------------------------------------------")
  playing = True
  rount = 0
  while playing:
    if rount % 2 == 0:
      current_robot = robot_one
      enemy_robot = robot_two
    else:
      current_robot = robot_two 
      enemy_robot = robot_one
 
    current_robot.print_status()
    store_on(current_robot)
    print("What part should I use to attack?")
    part_to_use = int(input("Choose a number part: "))

    enemy_robot.print_status()
    print("Which part to the enemy should we attack?")
    part_to_attack = int(input("Choose a enemy part to attack: "))
    print("-----------------------------------------------------------------\n")

    current_robot.add_coins(part_to_attack)
    current_robot.attack(enemy_robot, part_to_use, part_to_attack)
    rount += 1
    
    if enemy_robot.is_no() or enemy_robot.is_there_available_parts() :
      playing=False
      enemy_robot.print_status()
      print("Congratulation, you won")
      print(current_robot.name)

play()
```
